```python
df['strDrink'] = df['strDrink'].str.lower()
df['strIngredient1'] = df['strIngredient1'].str.lower()
df['strIngredient2'] = df['strIngredient2'].str.lower()
df['strIngredient3'] = df['strIngredient3'].str.lower()
df['strIngredient4'] = df['strIngredient4'].str.lower()
df['strIngredient5'] = df['strIngredient5'].str.lower()
df['strIngredient6'] = df['strIngredient6'].str.lower()
df['strIngredient7'] = df['strIngredient7'].str.lower()
df['strIngredient8'] = df['strIngredient8'].str.lower()
df['strIngredient9'] = df['strIngredient9'].str.lower()
df['strIngredient10'] = df['strIngredient10'].str.lower()
df['strIngredient11'] = df['strIngredient11'].str.lower()
df['strIngredient12'] = df['strIngredient12'].str.lower()
```

```python
# combine all strings:
df2 = df.drop(['Unnamed: 0','dateModified','idDrink','strAlcoholic','strCategory','strDrinkThumb','strGlass','strIBA','strIngredient13','strIngredient14','strIngredient15','strInstructions','strMeasure1','strMeasure2','strMeasure3','strMeasure4','strMeasure5','strMeasure6','strMeasure7','strMeasure8','strMeasure9','strMeasure10','strMeasure11','strMeasure12','strMeasure13','strMeasure14','strMeasure15','strVideo'],axis=1)

df2['data'] = df2[df2.columns[1:]].apply(
    lambda x: ' '.join(x.dropna().astype(str)),
    axis=1
)

print(df2['data'].head())
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
vectorized = vectorizer.fit_transform(df2['data'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarities = cosine_similarity(vectorized)
```

```python
print(similarities)
```

```python
df = pd.DataFrame(similarities, columns=df['strDrink'], index=df['strDrink']).reset_index()

df.head()
```

```python
input_cocktail = 'cosmopolitan'
recommendations = pd.DataFrame(df.nlargest(11,input_cocktail)['strDrink'])
recommendations = recommendations[recommendations['strDrink']!=input_cocktail]
print(recommendations)
```

```python
import pickle
pickle.dump(df.to_dict(),open('cocktail_dict.pkl','wb'))
pickle.dump(df,open('cocktail_similarity.pkl','wb'))
```
